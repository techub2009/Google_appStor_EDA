## üìù Note:
The values `Varies with device`, and they don‚Äôt make much sense. It appears that either the data is missing or incorrectly collected.

So, I will treat such values as missing (NaN).

Most of the data in the Size column is in MBs, while a few are in KBs.

To clean this column, we need to remove the letters ‚ÄòM‚Äô and ‚ÄòK‚Äô, and convert the values into a consistent format (preferably MBs).
For example, to convert KB into MB, we divide the number by 1024.

```python
df['Size']= df['Size'].str.replace(',','')
```

```python
# let's make a function that will remove M and k from the values and make it float data type, and also convert kbs into mbs

def Convert_m_k(size):
    if isinstance(size, str): # To check if size in string

        if 'M' in size: # to check M and remove it
            return float(size.replace('M','').strip())
        elif 'K' in size: # To check if size in kbs then remove k and convert it in mbs
            return float(size.replace('K','').strip()) / 1024
    else:
        return size # if none of above lye then provide us size

# apply function 
df['Size'] = df['Size'].apply(Convert_m_k) 

# let's convert data type form object to float
df['Size'] = pd.to_numeric(df['Size'], errors='coerce')

print(df['Size']. head(5))
    
```

```python
# let's apply IQR method to detect outlier from Rating column 
Q1 = df['Rating'].quantile(0.25)
Q3 = df['Rating'].quantile(0.75)
IQR = Q3 - Q1
print('This is the interquartile range',IQR)

# define limits
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
print('The lower limit is',lower_limit)
print('The upper limit is',upper_limit)

# detect outliers
outliers = df[(df['Rating'] < lower_limit) | (df['Rating'] > upper_limit)]
print('The outliers are',outliers)

# remove outliers
# df = df[(df['Rating'] >= lower_limit) & (df['Rating'] <= upper_limit)]



```
